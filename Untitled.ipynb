```python
from keras.datasets import cifar10
import numpy as np
import requests
import json
import urllib.parse

np.set_printoptions(threshold=np.inf)
```

```python
(trainX, trainY), (testX, testY) = cifar10.load_data()
```

```python
a = trainX[1]/255.0
a.shape
```

```python
b = a.reshape(-1)
c = b.tolist()
dct = {'w': c}
dct
```

```python
json_data = json.dumps(dct)
```

```python
encoded_data = urllib.parse.quote(json_data)
```

```python
# Define the URL to send the request to
url = 'http://127.0.0.1:5000/model-input'

# Set the headers to indicate that we are sending JSON data
headers = {'Content-Type': 'application/json'}

# Send the request with the JSON data in the request body
response = requests.post(url, data=json_data, headers=headers)

# Check the response status code to make sure the request was successful
if response.status_code == requests.codes.ok:
    print('Request successful')
else:
    print('Request failed')
```
